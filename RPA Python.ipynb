## Desafio

- Vamos automatizar a extração de informações de um sistema e envio de um relatório por e-mail
- No nosso caso, para todo mundo conseguir fazer o mesmo programa, o nosso "sistema" vai ser o Gmail, mas o mesmo processo pode ser feito com qualquer programa do seu computador e qualquer sistema
    - Passo 1: Entrar no sistema (entrar no Gmail)
    - Passo 2: Entrar em uma aba específica do sistema onde tem o nosso relatório (Aba Contatos)
    - Passo 3: Exportar o Relatório (Exportar Contatos)
    - Passo 4: Pegar o relatório exportado, tratar e pegar as informações que queremos
    - Passo 5: Preencher/Atualizar informações do sistema (No nosso caso, criar um e-mail e enviar)

- Link com um resumo dos principais comandos: https://pyautogui.readthedocs.io/en/latest/quickstart.html

```python

# OBS: 
# pyautogui.write() -> escreve
# pyautogui.click -> clica
# pyautogui.locateOnScreen -> identifica uma imagem na sua tela
# pyautogui.hotkey -> usa atalhos do teclado (combinação de teclas)
# pyautogui.press -> aperta um botão do teclado
# print(pyautogui.KEYBOARD_KEYS)

# O parâmetro confidence=0.8 ou 0.9 (que exige a instalação do opencv-python) permite que ele ignore essas pequenas diferenças de cores e pixels.

import pyautogui
import time

# Configurações iniciais
pyautogui.PAUSE = 1
pyautogui.FAILSAFE = True # Se você jogar o mouse no canto da tela, o código para

pyautogui.alert('O código vai começar. Não mexa em nada enquanto o código tiver rodando.')

# 1. Abrir o Chrome
pyautogui.press('win')
pyautogui.write("chrome")
pyautogui.press('enter')
time.sleep(2) # Espera o navegador abrir

# 2. Localizar Login (Usando try/except para não travar se não achar)
try:
    # Usei confidence=0.8 para facilitar a busca (requer: pip install opencv-python)
    res = pyautogui.locateCenterOnScreen('Foto01-Login Victor Menezes.png', confidence=0.8)
    if res:
        pyautogui.click(res)
except:
    print("Aviso: Foto01 não encontrada, seguindo para o próximo passo.")

# 3. Ir para o Gmail
pyautogui.write('https://gmail.com') 
pyautogui.press('enter')
time.sleep(2)

# 5. Clicar no Menu (Pontinhos)
foto_pontos = pyautogui.locateCenterOnScreen('Foto03-Logo Pontinhos.png', confidence=0.8)
if foto_pontos:
    pyautogui.click(foto_pontos)

# 6. Clicar em Contatos
time.sleep(1) # Esperar o menu abrir
try:
    foto_contatos = pyautogui.locateCenterOnScreen('Foto04-Contatos.png', confidence=0.8)
    pyautogui.click(foto_contatos)
except:
    print("Erro: Não encontrei o ícone de Contatos. Verifique se o menu abriu corretamente.")

# 7. Clicar no primeiro importar
time.sleep(1) # Esperar o menu abrir
try:
    foto_importar = pyautogui.locateCenterOnScreen('Foto05-Importar.png', confidence=0.8)
    pyautogui.click(foto_importar)
except:
    print("Erro: Não encontrei o ícone de Importar.")

# 8. Clicar no exportar para download

# 8.1 Esperar carregar o Exportar Contatos
while not pyautogui.locateOnScreen('Foto05.1-Exportar Contatos.png', confidence=0.8):
    time.sleep(1)
pyautogui.click(1097, 753) #Tela Cheia!

# try:
#     x, y, largura, altura = pyautogui.locateOnScreen("Foto05.2-Exportar.png", confidence=0.8)
#     pyautogui.click(x + largura/2, y + altura/2)
# except:
#     print("Erro: Não encontrei o ícone de Exportar para confirmar.")
```

```python
import pandas as pd
import pyperclip  # Copiar determinado texto de forma correta (Com pontuação e tudo)

df = pd.read_csv(r'C:\Users\victo\Downloads\contacts.csv')
df = df.dropna(axis=1) # Exclui colunar vazias
#display(df)

# 9. Seleciona a hotkey PgUP, para voltar a pagina de email
pyautogui.hotkey('ctrl', 'pgup') 

# 10. Escrever quantos emails existem no arquivo csv. 
for email in df['E-mail 1 - Value']:
    time.sleep(1)
    
    #10.1 Clica na foto compor
    foto_compor = pyautogui.locateCenterOnScreen("Foto06-Compor.png", confidence=0.8) 
    pyautogui.click(foto_compor)

    print('Erro: Não foi encontrado o ícone de Compor')

    # 10.2 Escrever o email
    pyautogui.write('victor_menezes@hotmail.com')
    pyautogui.press('enter')

    #10.3 Tab para pular para o assunto
    pyautogui.press('tab')
    pyautogui.write('Escrevendo um email teste')

    #10.4 Tab para o corpo do email
    pyautogui.press('tab')
    texto = '''Prezados, os dados dos contatos foram atualizados com sucesso no sistema hoje. 
    Segue a lista para conferência.'''
    pyperclip.copy(texto) # Usado para copiar a variável texto!
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.hotkey('ctrl', 'enter')
```
